```python
reg6.summary()
```

```python
from statsmodels.iolib.summary2 import summary_col
```

```python
regfinal = smf.ols('Motto_Likelihood ~ Health_Benefit + Tea + SparklingWater + Ingredients_Quality + Brand_Reputation + Price', data = pj).fit()
```

```python
regfinal.summary()
```

```python
summary_col([reg1, reg2, reg3, reg4, reg5, reg6], model_names=['Main', 'Tea', 'SparklingWater', 'Ingredients_Quality', 'Brand_Reputation', 'Price'], regressor_order=['R-squared', 'Intercept', 'Health_Benefit', 'Tea', 'SparklingWater', 'Ingredients_Quality', 'Brand_Reputation', 'Price'])
```

```python
summ = summary_col([reg1, regfinal], model_names=['Main', 'Control'], regressor_order=['R-squared','Intercept', 'Health_Benefit', 'Ingredients_Quality', 'Brand_Reputation', 'Price', 'SparklingWater', 'Tea'])
```
