```python
#ADDING CONFOUNDING VARIABLES: Tea 
reg2 = smf.ols('Motto_Price ~ Ingredients_Quality + Tea', data = pj).fit()
reg2.params
#Interpretation:  for every increase in tier of importance of ingredients quality when controlling for known confounding variables, 
#users on average willing to purchase Motto at 17 cents higher
```

```python
from statsmodels.iolib.summary2 import summary_col

summ = summary_col([reg1, reg2], model_names=['Main', 'Control'], regressor_order=['R-squared','Intercept'])
summ
```
